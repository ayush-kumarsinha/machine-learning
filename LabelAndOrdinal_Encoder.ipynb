Label And Ordinal Encoder

```python
import pandas as pd
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# create a simple dataframe with a categorical variable

df = pd.DataFrame({
    'color':['read', 'blue', 'green', 'red', 'blue']
})
```

```python
df
```

```python
#create an instance of LabeLencoder
encoder=LabelEncoder()
```

```python
encoder.fit_transform(df['color'])
```

```python
## ordinal Encoding

import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

# create a simple dataframe with an ordinal variable
df = pd.DataFrame({
    'size': ['small', 'medium', 'large', 'medium', 'small', 'large']
})
```

```python
## Create an instance of the OrdinalEncoder class and fit_transform
encoder=OrdinalEncoder(categories=[['small', 'medium', 'large']])
```

```python
encoder.fit_transform(df[['size']])
```
